```python
import requests

# URL ของ JSON ที่เก็บข้อมูลหลักสูตร
json_url = 'https://my-tcas.s3.ap-southeast-1.amazonaws.com/mytcas/courses.json'

# ส่งคำขอไปยัง API
response = requests.get(json_url)

# ตรวจสอบสถานะการตอบกลับ
if response.status_code == 200:
    data = response.json()  # แปลงข้อมูล JSON เป็น Python dict

    # กรองข้อมูลที่เกี่ยวข้องกับวิศวกรรมคอมพิวเตอร์
    count = 0  # ตัวแปรสำหรับนับจำนวนหลักสูตร
    for course in data:
        # ตรวจสอบคำว่า 'วิศวกรรมคอมพิวเตอร์' ในหลายๆ field
        program_name = course.get('program_name_th', '').lower()
        field_name = course.get('field_name_th', '').lower()
        faculty_name = course.get('faculty_name_th', '').lower()

        if ('วิศวกรรมคอมพิวเตอร์' in program_name or 
            'วิศวกรรมคอมพิวเตอร์' in field_name or 
            'วิศวกรรมคอมพิวเตอร์' in faculty_name):
            count += 1
            # แสดงผลข้อมูลที่เกี่ยวข้อง
            program = course.get('program_name_th', 'ไม่พบชื่อหลักสูตร')
            university = course.get('university_name_th', 'ไม่พบชื่อมหาวิทยาลัย')
            faculty = course.get('faculty_name_th', 'ไม่พบคณะ')
            cost = course.get('cost', 'ไม่พบข้อมูลค่าใช้จ่าย')

            # พิมพ์ข้อมูลที่กรองออกมา
            print(f"หลักสูตร: {program}")
            print(f"มหาวิทยาลัย: {university}")
            print(f"คณะ: {faculty}")
            print(f"ค่าใช้จ่าย: {cost}")
            print("-" * 50)

    # แสดงจำนวนหลักสูตรที่เกี่ยวข้องกับวิศวกรรมคอมพิวเตอร์ทั้งหมด
    print(f"จำนวนหลักสูตรวิศวกรรมคอมพิวเตอร์ทั้งหมด: {count} หลักสูตร")

else:
    print(f"ไม่สามารถดึงข้อมูลได้, สถานะ: {response.status_code}")
```

```python
import re
import pandas as pd

# แบ่งข้อมูลตามแต่ละหลักสูตร
courses = raw_data.strip().split("--------------------------------------------------")

# คำสำคัญที่เราสนใจ
keywords = {
    "ตลอดหลักสูตร": re.compile(r'ตลอดหลักสูตร'),
    "ต่อภาคเรียน": re.compile(r'ต่อภาคเรียน'),
    "ภาคการศึกษาแรก": re.compile(r'ภาคการศึกษาแรก'),
    "ร้อน": re.compile(r'ภาคฤดูร้อน'),
    "เทอม": re.compile(r'(เทอม|เทอมละ)'),
    "ภาคการศึกษา": re.compile(r'ภาคการศึกษา'),
    "ต่อภาค": re.compile(r'ต่อภาค'),
    "ภาคเรียน": re.compile(r'ภาคเรียน'),
    "เทอมเหมาจ่าย": re.compile(r'เทอมเหมาจ่าย'),
    "แรก": re.compile(r'(เทอมแรก|ภาคแรก|ครั้งแรก)')
}

results = []

for course in courses:
    lines = course.strip().split('\n')
    
    course_info = {
        "university": "",
        "program": "",
        "fee_numbers": [],
        "matched_keywords": set(),
        "original_text": course  # เพิ่มการเก็บต้นฉบับไว้
    }
    
    for line in lines:
        if line.startswith("มหาวิทยาลัย:"):
            course_info["university"] = line.replace("มหาวิทยาลัย:", "").strip()
        elif line.startswith("หลักสูตร:"):
            course_info["program"] = line.replace("หลักสูตร:", "").strip()
        elif any(line.startswith(prefix) for prefix in ["ค่าใช้จ่าย:", "ค่าธรรมเนียม", "อัตราค่าเล่าเรียน", "อัตราค่าบำรุงการศึกษา"]):
            cleaned_text = line.replace(',', '').strip()

            # หาเลขจำนวนเต็ม ≥ 5 หลัก
            numbers = re.findall(r'\b\d{5,}\b', cleaned_text)

            # ตรวจสอบว่า "ภาคการศึกษา" หรือ "ตลอดหลักสูตร" มีในข้อความหรือไม่
            has_semester_keyword = keywords["ภาคการศึกษา"].search(cleaned_text.lower())
            has_all_courses_keyword = keywords["ตลอดหลักสูตร"].search(cleaned_text.lower())
            has_term = keywords["เทอม"].search(cleaned_text.lower())
            has_first_term = keywords["ภาคการศึกษาแรก"].search(cleaned_text.lower())
            has_interm = keywords["ต่อภาคเรียน"].search(cleaned_text.lower())

            fee_numbers = []

            for num in numbers:
                numeric_value = int(num)
                
                # ถ้ามี "ตลอดหลักสูตร" ให้หาค่าใกล้เคียงที่สุดแล้วหาร 8
                if has_all_courses_keyword:
                    # ตรวจสอบหาค่าที่ใกล้เคียงสุดจากค่าธรรมเนียมในข้อมูล
                    if numeric_value >= 100000:  # ถ้ามีค่ามากกว่าหรือเท่ากับ 100,000
                        course_info["fee_numbers"].append(str(numeric_value // 8))  # หารด้วย 8 เพื่อให้ได้ค่าเทอม
                # ถ้าค่าธรรมเนียมต่ำกว่า 30,000 ก็ใช้ค่าเทอมตามปกติ
                elif numeric_value < 30000:
                    fee_numbers.append(str(numeric_value))
                else:
                    # หากพบ "ตลอดหลักสูตร" และ "ต่อภาคเรียน" คำนวณเป็นค่าใช้จ่ายต่อภาคเรียน
                    if has_all_courses_keyword and has_interm:
                        # ถ้าพบทั้ง "ตลอดหลักสูตร" และ "ต่อภาคเรียน" ใช้ค่า 14000 โดยตรง
                        course_info["fee_numbers"].append("14000")
                    elif has_semester_keyword or has_term or has_interm:
                        course_info["fee_numbers"].append(str(numeric_value))
                    else:
                        # ถ้าไม่พบคำสำคัญ "ภาคการศึกษา" ให้คำนวณเป็นค่าเทอม
                        if not has_semester_keyword:
                            course_info["fee_numbers"].append(str(numeric_value // 8))  # แบ่งค่าเทอมเป็น 8

                # ตรวจจับคำสำคัญ
                for keyword, pattern in keywords.items():
                    if pattern.search(cleaned_text.lower()):
                        course_info["matched_keywords"].add(keyword)

    # เก็บเฉพาะรายการที่มีค่า fee_numbers
    if course_info["fee_numbers"]:
        results.append(course_info)

# สร้าง DataFrame จากผลลัพธ์
df = pd.DataFrame(results, columns=["university", "program", "fee_numbers", "matched_keywords", "original_text"])

# บันทึกข้อมูลเป็น Excel
df.to_excel("data_clean_regex.xlsx", index=False, engine="openpyxl")
df.to_csv("data_clean_regex.csv", index=False, encoding="utf-8-sig")

print("Exported to 'data_clean_regex.xlsx'.")
print(df)
```
